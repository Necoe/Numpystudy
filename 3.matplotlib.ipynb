```python
import matplotlib.pyplot as plt
import numpy as np

# 可选：中文乱码问题（需要本地有相应中文字体）
plt.rcParams['font.sans-serif'] = ['SimHei']  # 显示中文
plt.rcParams['axes.unicode_minus'] = False   # 正常显示负号


# 折线图
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='sin(x)', color='blue', linestyle='--')
plt.title('正弦函数折线图')
plt.xlabel('x轴')
plt.ylabel('y轴')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 散点图
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='red', marker='o')
plt.title('随机散点图')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# 条形图
labels = ['A', 'B', 'C', 'D']
values = [10, 24, 36, 18]

plt.bar(labels, values, color='orange')
plt.title('简单柱状图')
plt.ylabel('数值')
plt.show()
```

```python
# 直方图
data = np.random.randn(1000)

plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('正态分布直方图')
plt.xlabel('值')
plt.ylabel('频数')
plt.show()
```

```python
# 饼图
labels = ['苹果', '香蕉', '葡萄', '橘子']
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # 突出显示第二块

plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%', shadow=True)
plt.title('水果销售比例')
plt.axis('equal')  # 保证饼图为圆
plt.show()
```

```python
# 多子图
x = np.linspace(0, 2 * np.pi, 100)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(x, np.sin(x))
plt.title('sin(x)')

plt.subplot(1, 2, 2)
plt.plot(x, np.cos(x))
plt.title('cos(x)')

plt.tight_layout()
plt.show()
```

```python
# 图像保存
x = np.linspace(0, 10, 100)
y = np.exp(-x / 3) * np.sin(x)

plt.plot(x, y)
plt.title('保存图像')
plt.savefig('plot_example.png', dpi=300)
plt.show()
```
